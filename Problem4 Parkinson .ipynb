```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import MinMaxScaler,StandardScaler,Normalizer,LabelEncoder,PolynomialFeatures
from sklearn.model_selection import train_test_split,cross_val_predict 
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC 
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
df = pd.read_csv('parkinson.csv')
df.head()
```

```python
minmax=MinMaxScaler()
X =df.drop(['status','name'],axis='columns')
y= df['status']
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3)
```

```python
X_train =minmax.fit_transform(X_train)
X_test = minmax.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(random_state=16)
logreg.fit(X_train,y_train)
logreg.score(X_test,y_test)
```

```python
y_pred=logreg.predict(X_test)
cnf_matrix = confusion_matrix(y_pred,y_test)
plt.figure(figsize=(10,7))
sns.heatmap(cnf_matrix,annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
print('-'*80)
print('Logistian regression accuracy:', accuracy_score(y_test, y_pred ))
print('-'*80)
```

```python
from sklearn import metrics
```

```python
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()  
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
dtc= DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_pred_dtc=dtc.predict(X_test)
cnf_matrix = metrics.confusion_matrix(y_test,y_pred_dtc)
plt.figure(figsize=(10,7))
sns.heatmap(cnf_matrix,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
print('-'*80)
print('Decision Tree accuracy:', accuracy_score(y_test, y_pred_dtc ))
print('-'*80)
fig=plt.figure(figsize=(25,20))


```

```python
fig=plt.figure(figsize=(25,20))
tree = plot_tree(dtc, feature_names = [ c for c in df.drop(['name'],axis=1).columns], class_names=['0','1'], filled=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
RM= RandomForestClassifier(n_estimators=20)
RM.fit(X_train,y_train)
RM.score(X_test,y_test)
y_pred_rm= RM.predict(X_test)
cm= metrics.confusion_matrix(y_test,y_pred_rm)
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
print('-'*80)
print('Random forest accuracy:', accuracy_score(y_test, y_pred_rm ))
print('-'*80)
```

```python
from sklearn.svm import SVC

SVM = SVC(probability=True, kernel = 'linear')
SVM.fit(X_train,y_train)
SVM.score(X_test, y_test)
y_pred_svm= SVM.predict(X_test)
cm=metrics.confusion_matrix(y_pred_svm,y_test)
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot = True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
print('-'*80)
print('SVM accuracy:', accuracy_score(y_test, y_pred_svm ))
print('-'*80)

```

```python
from sklearn.naive_bayes import GaussianNB
nby=GaussianNB()
nby.fit(X_train,y_train)
nby.score(X_test,y_test)
y_pred_nb=nby.predict(X_test)
cm=metrics.confusion_matrix(y_pred_nb,y_test)
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
print('-'*80)
print('Naive bayes classifier accuracy:',accuracy_score(y_test, y_pred_nb))
print('-'*80)
```
